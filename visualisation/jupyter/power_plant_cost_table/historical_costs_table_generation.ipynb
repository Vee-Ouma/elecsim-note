```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
historical_costs = pd.read_csv('~/Documents/PhD/Reports/6. ELECSIM/Paper/visualisation/jupyter/power_plant_cost_table/historical_costs.csv')

historical_costs
```

```python
historical_costs_diesel = historical_costs.copy()
historical_costs_diesel['Type'] = historical_costs_diesel['Type'].str.replace("Recip_diesel","Reciprocating Engine (Diesel)")

historical_costs_gas = historical_costs_diesel.copy()
historical_costs_gas['Type'] = historical_costs_diesel['Type'].str.replace("Recip_gas","Reciprocating Engine (Gas)")

# historical_costs_gas['Year'] = historical_costs_diesel['Year'].astype(str)
historical_costs_gas['Year'] = historical_costs_diesel['Year'].astype("int")
historical_costs_gas

historical_costs_gas = historical_costs_gas.rename(index=str, columns={'Plant_Size':'Capacity'})

```

```python
cost_index = historical_costs_gas.set_index(['Type', 'Capacity', 'Year'])
cost_index
```

```python
cost_sorted = cost_index.sort_values(['Type','Capacity','Year'])
cost_sorted
```

```python
cost_ordered = cost_sorted[['efficiency','operating_period','pre_dev_period','construction_period','pre_dev_cost_per_mw','construction_cost_per_mw','infrastructure','fixed_o_and_m_per_mw', 'variable_o_and_m_per_mwh', 'insurance_cost_per_mw','connection_cost_per_mw']]
cost_ordered
```

```python
columns_to_round = ['pre_dev_cost_per_mw','construction_cost_per_mw','infrastructure','fixed_o_and_m_per_mw','insurance_cost_per_mw','connection_cost_per_mw']
cost_rounded = cost_ordered.copy()
cost_rounded[columns_to_round] = cost_ordered[columns_to_round].round()

columns_remove_trailing_0 = ['operating_period','pre_dev_period','construction_period','pre_dev_cost_per_mw','construction_cost_per_mw','infrastructure','fixed_o_and_m_per_mw','variable_o_and_m_per_mwh','insurance_cost_per_mw','connection_cost_per_mw']

cost_rounded[columns_remove_trailing_0] = cost_ordered[columns_remove_trailing_0].astype(np.int64)

cost_rounded
```

```python
cost_comma = cost_rounded.copy()
cost_comma["pre_dev_cost_per_mw"] = cost_rounded["pre_dev_cost_per_mw"].map(lambda x: "{:,}".format(x))
cost_comma["construction_cost_per_mw"] = cost_rounded["construction_cost_per_mw"].map(lambda x: "{:,}".format(x))
cost_comma["infrastructure"] = cost_rounded["infrastructure"].map(lambda x: "{:,}".format(x))
cost_comma["fixed_o_and_m_per_mw"] = cost_rounded["fixed_o_and_m_per_mw"].map(lambda x: "{:,}".format(x))
cost_comma["connection_cost_per_mw"] = cost_rounded["connection_cost_per_mw"].map(lambda x: "{:,}".format(x))
cost_comma["insurance_cost_per_mw"] = cost_rounded["insurance_cost_per_mw"].map(lambda x: "{:,}".format(x))
cost_comma
```

```python
# formalism = cost_comma.rename(index=str, columns = {"efficiency":"$\eta$", 'operating_period':"$OP$", "pre_dev_period":"$P_D$", 'construction_period':'$C_D$','pre_dev_cost_per_mw':"$P_C$", 'construction_cost_per_mw':'$C_C$', 'infrastructure':'$I_C$','fixed_o_and_m_per_mw':'$F_C$','variable_o_and_m_per_mwh':'$V_C$',"insurance_cost_per_mw":'$In_C$',"connection_cost_per_mw":'$Con_C$'})
formalism = cost_comma
formalism
```
