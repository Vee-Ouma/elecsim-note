```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
power_plant_costs = pd.read_csv("/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/data/processed/power_plants/power_plant_costs/modern_power_plant_costs/power_plant_costs_with_simplified_type.csv")

power_plant_costs
```

```python
def get_columns_of_with_string(cost_dataframe, year):
    year_cols = [col for col in cost_dataframe.columns if year in col]
    return year_cols

def get_stub_of_costs(cost_dataframe):
    year_cols = [col.split(" _")[0]+" _" for col in cost_dataframe.columns if "2018" in col]
    return year_cols

def get_medium_stubs(column_names):
    medium_colums = [x for x in column_names if "Medium" in x]
    return medium_colums
    
get_columns_of_with_string(power_plant_costs, '2018')

get_medium_stubs(get_stub_of_costs(power_plant_costs))
```

```python
costs_to_display = power_plant_costs[['Plant_Type', 'Type','Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_with_string(power_plant_costs, '2018') + get_columns_of_with_string(power_plant_costs, '2020') + get_columns_of_with_string(power_plant_costs, '2025'))]
costs_to_display
# print(['Type', 'Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_specific_year(power_plant_costs, 2018)))

costs_to_display_diesel = costs_to_display.copy()
costs_to_display_diesel['Type'] = costs_to_display['Type'].str.replace("Recip_diesel","Recip. Engine (Diesel)")

costs_to_display_gas = costs_to_display_diesel.copy()
costs_to_display_gas['Type'] = costs_to_display_diesel['Type'].str.replace("Recip_gas","Recip. Engine (Gas)")

costs_to_display_gas
      
      
```

```python

columns_to_drop = ['Pre_1','Pre_2','Pre_3','Pre_4','Pre_5','Pre_6','Pre_7','Pre_8', 'Unnamed: 0', 'Constr_1','Constr_2','Constr_3','Constr_4','Constr_5','Constr_6','Constr_7','Constr_8','Average_Load_Factor']
```

```python
formated_costs = pd.wide_to_long(costs_to_display_gas, i=['Plant_Type','Plant_Size','Type'], j='year', stubnames=get_stub_of_costs(power_plant_costs))
formated_costs
```

```python
# formated_costs_simplified = formated_costs.drop(columns_to_drop, axis=1)
# formated_costs_simplified
```

```python
costs_dropna = formated_costs.dropna(subset=get_stub_of_costs(power_plant_costs), how="all")
costs_dropna
```

```python
costs_reset_index = costs_dropna.reset_index()
costs_reset_index = costs_reset_index.drop('Plant_Type', axis=1)
costs_reset_index
```

```python
costs_common_types = costs_reset_index[~costs_reset_index.Type.isin(['Tidal','Coal_Retro',"EfW","AD","ACT","Biomass","Conversions","Landfill","Sewage","Geothermal","Hydro_Store","Wave","CCGT_CCS","Pumped_storage", "OCGT_CCS"])]
costs_common_types_index = costs_common_types.set_index(['Type', "Plant_Size","year"])

costs_sorted = costs_common_types_index.sort_values(['Type','Plant_Size','year'])
costs_sorted

```

```python
medium_columns = [x for x in costs_sorted.columns if " _" and "Medium" in x]
medium_columns

costs_medium = costs_sorted[['Efficiency','Operating_Period','Pre_Dur','Constr_Dur']+medium_columns]
costs_medium
```

```python
costs_medium_ordered = costs_medium[['Efficiency','Operating_Period','Pre_Dur','Constr_Dur','Pre_dev_cost-Medium _','Constr_cost-Medium _', 'Infra_cost-Medium _','Fixed_cost-Medium _','Var_cost-Medium _','Insurance_cost-Medium _','Connect_system_cost-Medium _']]

costs_medium_ordered = costs_medium_ordered.rename(columns={col: col.split("-Medium ")[0] for col in costs_medium_ordered.columns})

costs_medium_ordered
```

```python
costs_dropped_duplicates = costs_medium_ordered.drop_duplicates()
costs_dropped_duplicates
```

```python
costs_reset_index_dropped = costs_dropped_duplicates.reset_index(['year'])
costs_reset_index_dropped
def years_up(group):
    if all(group.year == 2018):
        return group.replace(2018, '2018/20/25')
    elif all(group.year == 2020):
        return group.replace(2020, '2020/25')
    else:
        return group
#     print(all(group.year == 2018))
# costs_reset_index_dropped
costs_fixed_year = costs_reset_index_dropped.groupby("Type").apply(lambda x: years_up(x))
costs_fixed_year
costs_fixed_year_index = costs_fixed_year.set_index(["year"], append=True)
costs_fixed_year_index
```

```python
costs_removed_duplicated_indices = costs_fixed_year_index[~costs_fixed_year_index.index.duplicated(keep='first')]

costs_removed_duplicated_indices

```

```python
columns_remove_trailing_0 = ['Operating_Period','Pre_dev_cost','Constr_cost','Infra_cost','Fixed_cost','Var_cost','Insurance_cost','Connect_system_cost']

costs_removed_duplicated_indices[columns_remove_trailing_0] = costs_removed_duplicated_indices[columns_remove_trailing_0].astype(np.int64)
```

```python

# costs_removed_duplicated_indices = costs_removed_duplicated_indices.apply(lambda x: "{:,}".format(x['Constr_cost']), axis=1)

costs_removed_duplicated_indices["Pre_dev_cost"] = costs_removed_duplicated_indices["Pre_dev_cost"].map(lambda x: "{:,}".format(x))
costs_removed_duplicated_indices["Constr_cost"] = costs_removed_duplicated_indices["Constr_cost"].map(lambda x: "{:,}".format(x))
costs_removed_duplicated_indices["Infra_cost"] = costs_removed_duplicated_indices["Infra_cost"].map(lambda x: "{:,}".format(x))
costs_removed_duplicated_indices["Fixed_cost"] = costs_removed_duplicated_indices["Fixed_cost"].map(lambda x: "{:,}".format(x))
costs_removed_duplicated_indices["Connect_system_cost"] = costs_removed_duplicated_indices["Connect_system_cost"].map(lambda x: "{:,}".format(x))
costs_removed_duplicated_indices["Insurance_cost"] = costs_removed_duplicated_indices["Insurance_cost"].map(lambda x: "{:,}".format(x))


costs_removed_duplicated_indices
```

```python
formalism = costs_removed_duplicated_indices.rename(index=str, columns = {"Efficiency":"$\eta$", 'Operating_Period':"$OP$", "Pre_Dur":"$P_D$", 'Constr_Dur':'$C_D$','Pre_dev_cost':"$P_C$", 'Constr_cost':'$C_C$', 'Infra_cost':'$I_C$','Fixed_cost':'$F_C$','Var_cost':'$V_C$',"Insurance_cost":'$In_C$',"Connect_system_cost":'$Con_C$'})

formalism = formalism.reset_index()
formalism = formalism.rename(columns = {'Plant_Size':'Capacity', 'year':'Year'})
formalism.set_index(['Type', 'Capacity (MW)', "Year"])
```
