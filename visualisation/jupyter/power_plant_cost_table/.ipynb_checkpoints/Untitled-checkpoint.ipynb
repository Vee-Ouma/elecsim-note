```python
def get_columns_of_specific_year(cost_dataframe, year):
    year = str(year)
    year_cols = [col for col in cost_dataframe.columns if year in col]
    return year_cols
    
get_columns_of_specific_year(power_plant_costs, 2018)
```

```python
costs_to_display = power_plant_costs[['Type', 'Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_specific_year(power_plant_costs, 2018))]
costs_to_display
# print(['Type', 'Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_specific_year(power_plant_costs, 2018)))
 
      
      
      
      
```
