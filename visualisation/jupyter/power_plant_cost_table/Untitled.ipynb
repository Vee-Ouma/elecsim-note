```python
def get_columns_of_with_string(cost_dataframe, year):
    year_cols = [col for col in cost_dataframe.columns if year in col]
    return year_cols

def get_stub_of_costs(cost_dataframe):
    year_cols = [col.split(" _")[0]+" _" for col in cost_dataframe.columns if "2018" in col]
    return year_cols
    
get_columns_of_with_string(power_plant_costs, '2018')

get_stub_of_costs(power_plant_costs)
```

```python
costs_to_display = power_plant_costs[['Plant_Type', 'Type','Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_with_string(power_plant_costs, '2018'))]
costs_to_display
# print(['Type', 'Plant_Size', 'Efficiency','Operating_Period','Constr_Dur','Pre_Dur'] + (get_columns_of_specific_year(power_plant_costs, 2018)))
 
      
      
      
      
```

```python

columns_to_drop = ['Pre_1','Pre_2','Pre_3','Pre_4','Pre_5','Pre_6','Pre_7','Pre_8', 'Unnamed: 0', 'Constr_1','Constr_2','Constr_3','Constr_4','Constr_5','Constr_6','Constr_7','Constr_8']
```

```python
formated_costs = pd.wide_to_long(power_plant_costs, i=['Plant_Type','Plant_Size','Type'], j='year', stubnames=get_stub_of_costs(power_plant_costs))
formated_costs_simplified = formated_costs.drop(columns_to_drop, axis=1)
formated_costs_simplified
```

```python
costs_dropna = formated_costs_simplified.dropna(subset=get_stub_of_costs(power_plant_costs), how="all")
costs_dropna
```

```python
index_number = pd.Series(list(range(len(costs_dropna))))

costs_reset_index = costs_dropna.reset_index()
costs_reset_index = costs_reset_index.drop('Plant_Type', axis=1)
costs_reset_index
```

```python
costs_common_types = costs_reset_index[~costs_reset_index.Type.isin(['Tidal','Coal_Retro',"EfW","AD","ACT","Biomass","Conversions","Landfill","Sewage","Geothermal","Hydro_Store","Wave","CCGT_CCS","Pumped_storage"])]
costs_common_types_index = costs_common_types.set_index(['Type', "Plant_Size","year"])

costs_common_types_index.sort_values(['Type','Plant_Size','year'])
```
